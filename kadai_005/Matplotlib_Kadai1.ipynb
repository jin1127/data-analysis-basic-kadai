```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2, 2, tight_layout=True)

# 折れ線
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

# sin関数
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
axes[1, 0].plot(x1, y1, linewidth=2.0)

axes[1, 0].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

#ヒストグラム
x2 = np.random .normal(15, 5, 2000)
axes[0, 1].hist(x2)

#散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1, 1].scatter(x3, y3)

plt.show()
```
